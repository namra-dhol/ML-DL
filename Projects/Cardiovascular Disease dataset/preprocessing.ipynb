```python
df2["BMI"] = (df2["weight"] / ((df2["height"] / 100) ** 2))
df2['BMI']
```

```python
swap_mask = df2['ap_hi'] < df2['ap_lo']
df2.loc[swap_mask, ['ap_hi', 'ap_lo']] = df2.loc[swap_mask, ['ap_lo', 'ap_hi']].values

df2['ap_hi'] = df2['ap_hi'].clip(lower=70, upper=240)
df2['ap_lo'] = df2['ap_lo'].clip(lower=40, upper=140)
```

```python
df2 = df2[
    (df2['ap_lo'] >= 40) & (df2['ap_lo'] <= 140) &
    (df2['ap_hi'] >= 70) & (df2['ap_hi'] <= 240)
]
```

```python
print(df2.columns.tolist())
```

```python
for col in ['BMI', 'ap_hi', 'ap_hi', 'BMI', 'age']:
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mask = (df2[col] < lower) | (df2[col] > upper)
    outliers = df2[mask]
    if not outliers.empty:
        df2[col] = np.where(df2[col] < lower, lower, df2[col])
        df2[col] = np.where(df2[col] > upper, upper, df2[col])
```
