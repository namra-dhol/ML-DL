```python
import numpy as np 

# output 1 is positive , 0 if negative
def step_function(x):
    if x>=0:
        return 1
    else:
        return 0


print("setup complete")
```

```python
# input 
x_or = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# traget 

y_or = np.array([0,1,1,1])
```

```python
weights = np.array([0,0])
bias = 0
lr = 0.5 # learning rate 

print(weights)
print(bias)
```

```python
#example input : (0,1)


input_sample = x_or[1]
target_value = y_or[1]

# 1 . weighted sum (Dot product)
z = np.dot(input_sample,weights) + bias 
print(f"Math ; ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {bias}")
print(f"Sum (z) : {z:.2f}")

# 2 activation 
prediction = step_function(z)
print(f"prediction : {prediction}")

#3 error 
error = target_value - prediction
print(f"error : {error}")
```

```python
epochs = 5
print("Traing")
for epoch in range(epochs):
    print(f"\nEPOCH->{epoch+1}")

    for i in range(4):
        x = x_or[i]
        y = y_or[i]
    
        #Predict
        z = np.dot(x,weights)+bias
        pred = step_function(z)
        error = y-pred
    
        #Learn (update weights)
        if error !=0:
            weights = weights+(lr*error*x)
            bias = bias+(lr*error*1) #bias is always one
            print(f"->MISTAKE ON {x}.Expected {y},got {pred}.")
            print(f"->UPDATED WEIGTHS:{weights},BIAS:{bias:.2f}")
        else:
            print(f" ✅ Correct on {x}")

print("\nFinal Weight:",weights)
print("Final Bias:",bias)
```
