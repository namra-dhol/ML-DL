# Decision Tree Classifier

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data = load_iris()
```

```python
data
```

```python
data.feature_names
```

# Importing the libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
```

# Divide the data into input and output

```python
data.target
```

```python
data.target_names
```

# Splitting the dataset into the Training set and Test set

```python
x = data['data']
```

```python
y = data['target']
```

# Fitting DecisionTreeClassifier   on dataset

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(f"Final shape of Training Data: {x_train.shape}")
print(F"final test of traing data :: {x_test.shape}")
print(f"Final shape of Training Data: {y_train.shape}")
print(F"final test of traing data :: {y_test.shape}")
```

```python
model = DecisionTreeClassifier(splitter="random")
```

```python
model.fit(x_train,y_train)
```

# Display Decision Tree

```python
from sklearn import tree
```

```python
%matplotlib qt
```

```python
tree.plot_tree(model)
```

# Predict the x_test 

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Import diabetes.csv dataset 

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

# Check the distribution of the target

```python
print(df['Outcome'].value_counts())
```

# Replace zeros with NaN for specific columns

```python
columns_to_replace = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
df[columns_to_replace] = df[columns_to_replace].replace(0,np.nan)
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Fill missing values with median

```python
for col in columns_to_replace:
    df[col] = df[col].fillna(df[col].mean())
```

```python
print(df.isnull().sum())
```

# Visualize Distributions

```python
sns.pairplot(df,hue='Outcome')
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.plot()
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({"Diabetic":1,"Non Diabetic" : 0 })
```

```python
plt.plot()
```

# Divide the data into input and output

```python
x = df.iloc[::,:-1:]
```

```python
y =  df['Outcome']
```

# Splitting the dataset into the Training set and Test set

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
print(f"Final shape of Training Data: {x_train.shape}")
print(F"final test of traing data :: {x_test.shape}")
print(f"Final shape of Training Data: {y_train.shape}")
print(F"final test of traing data :: {y_test.shape}")
```

# Create Model

```python
model =  DecisionTreeClassifier(splitter='random')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Fitting DecisionTreeClassifier   on dataset

# Display Decision Tree

```python
from sklearn import tree
%matplotlib qt
tree.plot_tree(model)
```

# Predict the x_test 

```python
y_predict = model.predict(x_test)
```

# Display Training Accuracy

```python
model.score(x_train,y_train)
```

# Display Test Accuracy

```python
model.score(x_test,y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_predict,)
cm
```
